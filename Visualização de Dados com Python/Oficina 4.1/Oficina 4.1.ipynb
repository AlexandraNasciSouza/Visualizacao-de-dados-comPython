```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style='darkgrid')
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
#Crie uma visualização que mostre a distribuição do valor da passagem (“fare”), de acordo com a cidade na qual o passageiro 
#embarcou ("embark_town"). Você pode escolher a técnica que julgar mais adequada;

plt.figure(figsize=(10,5))
plt.title('Distribuição do valor da passagem, de acordo com a cidade na qual o passageiro embarcou')
sns.violinplot(x='embark_town', y='fare', data=titanic)
plt.show()
```

```python
#Crie uma visualização que mostre as distribuições do valor da passagem, de acordo com a cidade de embarque, sexo e se o 
#passageiro sobreviveu ou não. Você pode escolher a técnica que julgar mais adequada. Que diferença você consegue notar nas 
#distribuições dos passageiros que sobreviveram em relação aos que não sobreviveram?
#R = Podemos ver que sobreviveram mais mulheres e também que a faixa de valor da pasagens da maioria dos sobriveventes se dá 
#entre 0 a 100.

plt.figure(figsize=(10,5))
plt.title('Distribuição do valor da passagem, de acordo com a cidade na qual o passageiro embarcou, sexo e se o passageiro sobreviveu ou não')
sns.catplot(x="embark_town", y="fare", hue = 'sex', col = 'survived', kind='swarm',data=titanic)
plt.show()
```
