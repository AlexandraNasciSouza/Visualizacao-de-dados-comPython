```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
filmes = pd.read_json('dadosoficina03.txt')
filmes
```

```python
#Crie uma visualizao que mostre o oramento de todos os filmes do conjunto de dados. 
#Qual o filme com maior oramento e com menor oramento? 
#R =  Pirates of the Caribbean: At World's End e The Twilight Saga: New Moon

plt.figure(figsize=(20,10))
plt.title('Oramento por filme')
sns.barplot(x="Budget_M", y="Film", color = 'Blue', estimator = sum, ci = None, data=filmes)
plt.show()
```

```python
#Crie uma visualizao que mostre a bilheteria de todos os filmes do conjunto de dados. 
#Qual o filme com maior bilheteria e com menor bilheteria?
#R= Avatar e Ratatouille

plt.figure(figsize=(20,10))
plt.title('Bilheteira filme')
sns.barplot(x="Worldwide_Gross_M", y="Film", color = 'orange', estimator = sum, ci = None, data=filmes)
plt.show()
```

```python
#Crie uma visualizao que mostre a quantidade de filmes por gnero em cada ano. Em que anos foram feitos menos filmes de ao? 
#E em que ano foram feitos menos filmes de animao?
#R = ao 2008 e 2010 e animao 2008 e 2011


plt.figure(figsize=(15,8))
plt.title('Filmes por gnero em cada ano')
sns.countplot(x="Genre", hue= 'Year', data=filmes)
plt.show()
```

```python
#Crie uma visualizao que mostre a mdia de oramento de filmes por gnero em cada ano. 
#Em que ano os filmes de ao tiveram maior mdia de oramento?
#E em que ano os filmes de animao tiveram a maior mdia de oramento?
#R = ao 2007 e animao 2008

plt.figure(figsize=(15,8))
plt.title('Mdia de oramento por gnero em cada ano')
sns.barplot(x="Year", y = 'Budget_M' , ci = None, hue= 'Genre', data=filmes)
plt.show()
```
