```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
jogos = pd.read_csv('dadosoficina04.txt')
jogos
```

```python
#Crie uma visualizaÃ§Ã£o que mostre a proporÃ§Ã£o de vendas global entre gÃªneros de jogos da empresa Nintendo;
jogos_nintendo = jogos[jogos['Publisher'] == 'Nintendo'].groupby('Genre').sum()
jogos_nintendo 
```

```python
jogos_nintendo = jogos[jogos['Publisher'] == 'Nintendo'].groupby('Genre').sum()
vendas = jogos_nintendo['Global_Sales']
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(15,5))
plt.bar(jogos_nintendo.index, vendas)
plt.show()
```

```python
#Crie uma visualizaÃ§Ã£o que mostre a proporÃ§Ã£o de quantidade de jogos por plataforma a partir do ano de 2015.
jogos_ano = jogos[jogos['Year'] >= 2015].groupby('Platform').count()
jogos_ano
```

```python
jogos_ano = jogos[jogos['Year'] >= 2015].groupby('Platform').count()
ano = jogos_ano['Year']
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(15,5))
plt.bar(jogos_ano.index, ano)
plt.show()
```
