```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style = "darkgrid")
```

```python
df = pd.read_csv('dadosoficina07.txt')
df.head(40)
```

```python
#Para o segundo e último passo, crie uma visualização que mostre a série temporal das vendas ao longo dos meses. Você pode, 
#por exemplo, se perguntar se existe alguma tendência ou alguma periodicidade, OK? 
#R= Que no dia primeiro de cada mês a vendas não são tão altas e que com o passar dos dias o valor ganho com as vendas sobem.

plt.figure(figsize=(20,5))
plt.title('')
sns.lineplot(x="Month", y="Sales", data=df)
plt.show()
```
